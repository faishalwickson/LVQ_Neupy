```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
# from keras.models import Sequential
# from keras.layers import Dense
```

```python
#input dataset
df = pd.read_csv('uts-sisdas.csv')
df
```

```python
#Hapus atribut no_obs
df = df.drop(columns='no_obs')
```

```python
#Konvert nilai kelas ke [0,2] sesuai format pada neupy
df.kelas[df.kelas == 1] = 0
df.kelas[df.kelas == 2] = 1
df.kelas[df.kelas == 3] = 2
```

```python
#mengubah format dataset
dataset = df.values
dataset
```

```python
# Pilih 3 kolom pertama untuk dijadikan sebagai atribut
X = dataset[:,0:3]
# Pilih 1 kolom terakhir sebagai label
y = dataset[:,3:4]
```

```python
#Lakukan normalisasi untuk meningkatkan performa dan akurasi
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
#Split dataset menjadi datatrain dan datatest dengan perbandingan 7:3
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
#melakukan training menggunakan neupy sebanyak 20 epochs dengan learning rate 0,01
lvqnet = algorithms.LVQ3(n_inputs=3, n_classes=3, step=0.01)
lvqnet.train(X_train, Y_train, epochs=20)
```

```python
#prediksi datatest
predicted = lvqnet.predict(X_test)
predicted
```

```python
#Evaluation
def print_metrics(labels, preds):
    print("Precision Score: {}".format(precision_score(labels,
           preds, average = 'weighted')))
    print("Recall Score: {}".format(recall_score(labels, preds,
           average = 'weighted')))
    print("Accuracy Score: {}".format(accuracy_score(labels,
           preds)))
    print("F1 Score: {}".format(f1_score(labels, preds, average =
           'weighted')))
```

```python
#Membandingkan datatest dengan prediksi untuk mendapatkan metrics
print_metrics(Y_test, predicted)
```
